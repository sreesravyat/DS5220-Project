```python
import pandas as pd
import joblib
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from IPython.display import display, IFrame
import ipywidgets as widgets

preprocessor = joblib.load('preprocessor.pkl')
best_model = joblib.load('best_model.pkl')

def predict_heart_disease(input_data):
    input_df = pd.DataFrame([input_data])
    input_preprocessed = preprocessor.transform(input_df)
    prediction = best_model.predict(input_preprocessed)
    return 'Heart Disease' if prediction[0] == 1 else 'No Heart Disease'

def prediction_reasons(input_data):
    reasons = []
    disease_types = set()
    
    if input_data['age'] > 50:
        reasons.append("Age is over 50, which is a risk factor.")
        disease_types.add("Coronary Artery Disease")
    if input_data['ap_hi'] > 140:
        reasons.append("Systolic blood pressure is high.")
        disease_types.add("Hypertension")
    if input_data['cholesterol'] > 1:
        reasons.append("Cholesterol level is above normal.")
        disease_types.add("Hypercholesterolemia")
    if input_data['gluc'] > 1:
        reasons.append("Glucose level is above normal.")
        disease_types.add("Diabetes")
    if input_data['smoke'] == 1:
        reasons.append("Smoking is a risk factor.")
        disease_types.add("Coronary Artery Disease")
    if input_data['alco'] == 1:
        reasons.append("Alcohol consumption is a risk factor.")
        disease_types.add("Cardiomyopathy")
    if input_data['active'] == 0:
        reasons.append("Lack of physical activity is a risk factor.")
        disease_types.add("Coronary Artery Disease")
    
    return reasons, disease_types

def save_data(data, filename='predictions.csv'):
    df = pd.DataFrame([data])
    try:
        existing_df = pd.read_csv(filename)
        df = pd.concat([existing_df, df], ignore_index=True)
    except FileNotFoundError:
        pass
    df.to_csv(filename, index=False)

name = widgets.Text(description="Name:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
identification = widgets.Text(description="ID:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
age = widgets.IntText(description="Age:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
gender = widgets.Dropdown(options=[('Male', 1), ('Female', 2)], description="Gender:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
height = widgets.IntText(description="Height (cm):", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
weight = widgets.IntText(description="Weight (kg):", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
ap_hi = widgets.IntText(description="Systolic Blood Pressure (mmHg):", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
ap_lo = widgets.IntText(description="Diastolic Blood Pressure (mmHg):", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
cholesterol = widgets.Dropdown(options=[("Normal", 1), ("Above Normal", 2), ("Well Above Normal", 3)], description="Cholesterol:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
gluc = widgets.Dropdown(options=[("Normal", 1), ("Above Normal", 2), ("Well Above Normal", 3)], description="Glucose:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
smoke = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Do you smoke?:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
alco = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Do you consume alcohol?:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))
active = widgets.Dropdown(options=[("No", 0), ("Yes", 1)], description="Are you physically active?:", style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))

def on_submit(b):
    input_data = {
        'name': name.value,
        'identification': identification.value,
        'age': age.value,
        'gender': gender.value,
        'height': height.value,
        'weight': weight.value,
        'ap_hi': ap_hi.value,
        'ap_lo': ap_lo.value,
        'cholesterol': cholesterol.value,
        'gluc': gluc.value,
        'smoke': smoke.value,
        'alco': alco.value,
        'active': active.value
    }
    prediction = predict_heart_disease(input_data)
    reasons, disease_types = prediction_reasons(input_data)
    input_data['prediction'] = prediction
    input_data['reasons'] = ', '.join(reasons)
    input_data['disease_types'] = ', '.join(disease_types)
    save_data(input_data)
    
    with open("prediction_result.html", "w") as f:
        f.write(f"<h1>Prediction for {input_data['name']} (ID: {input_data['identification']}): {prediction}</h1>")
        f.write("<h2>Possible Heart Diseases:</h2>")
        f.write("<ul>")
        for disease in disease_types:
            f.write(f"<li>{disease}</li>")
        f.write("</ul>")
        f.write("<h2>Reasons:</h2>")
        f.write("<ul>")
        for reason in reasons:
            f.write(f"<li>{reason}</li>")
        f.write("</ul>")
    
    display(IFrame(src="prediction_result.html", width=600, height=400))

submit_button = widgets.Button(description="Predict", button_style='success', layout=widgets.Layout(width='400px'))
submit_button.on_click(on_submit)

result = widgets.Text(value='', description='Prediction:', disabled=True, style={'description_width': 'initial'}, layout=widgets.Layout(width='400px'))

form = widgets.VBox([
    name, identification, age, gender, height, weight, ap_hi, ap_lo, cholesterol, gluc, smoke, alco, active, submit_button, result
])

display(form)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report, roc_curve, auc

new_data = pd.read_csv("C:/Users/jagad/Downloads/heart_disease_uci.csv")
new_data.head()

column_mapping = {
    'age': 'age',
    'sex': 'gender',
    'cp': 'cp',
    'trestbps': 'ap_hi',
    'chol': 'cholesterol',
    'fbs': 'fbs',
    'restecg': 'restecg',
    'thalch': 'thalach',
    'exang': 'exang',
    'oldpeak': 'oldpeak',
    'slope': 'slope',
    'ca': 'ca',
    'thal': 'thal',
    'num': 'target'
}

new_data.rename(columns=column_mapping, inplace=True)

new_data['fbs'].fillna(new_data['fbs'].mode()[0], inplace=True)
new_data['restecg'].fillna(new_data['restecg'].mode()[0], inplace=True)
new_data['exang'].fillna(new_data['exang'].mode()[0], inplace=True)
new_data['slope'].fillna(new_data['slope'].mode()[0], inplace=True)
new_data['thal'].fillna(new_data['thal'].mode()[0], inplace=True)
new_data['ca'].fillna(new_data['ca'].median(), inplace=True)

new_data['gender'] = new_data['gender'].map({'Male': 1, 'Female': 2})
new_data['cp'] = new_data['cp'].map({
    'typical angina': 0, 'atypical angina': 1, 
    'non-anginal': 2, 'asymptomatic': 3
})
new_data['fbs'] = new_data['fbs'].astype(int)
new_data['restecg'] = new_data['restecg'].map({
    'normal': 0, 'lv hypertrophy': 1
})
new_data['exang'] = new_data['exang'].astype(int)
new_data['slope'] = new_data['slope'].map({
    'upsloping': 0, 'flat': 1, 'downsloping': 2
})
new_data['thal'] = new_data['thal'].map({
    'normal': 1, 'fixed defect': 2, 'reversable defect': 3
})

features = ['age', 'gender', 'cp', 'ap_hi', 'cholesterol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
target = 'target'

numeric_features = ['age', 'ap_hi', 'cholesterol', 'thalach', 'oldpeak']
categorical_features = ['gender', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

X_train, X_test, y_train, y_test = train_test_split(new_data[features], new_data[target], test_size=0.2, random_state=42)

X_train = preprocessor.fit_transform(X_train)
X_test = preprocessor.transform(X_test)

joblib.dump(preprocessor, 'preprocessor_new.pkl')

models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'SVM': SVC(probability=True),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'Extra Trees': ExtraTreesClassifier(),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    'LightGBM': LGBMClassifier()
}

results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test) if hasattr(model, "predict_proba") else None
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro')
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    roc_auc = roc_auc_score(y_test, y_proba, average='macro', multi_class='ovo') if y_proba is not None and y_proba.ndim == 2 else None
    
    results[name] = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'roc_auc': roc_auc
    }
    
    print(f'{name} Results:')
    print(f'Accuracy: {accuracy}')
    print(f'Precision: {precision}')
    print(f'Recall: {recall}')
    print(f'F1-Score: {f1}')
    if roc_auc is not None:
        print(f'ROC-AUC: {roc_auc}')
    print('\n')
    print('Confusion Matrix:')
    print(confusion_matrix(y_test, y_pred))
    print('Classification Report:')
    print(classification_report(y_test, y_pred))
    
    if y_proba is not None and y_proba.ndim == 2:
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1], pos_label=1)
        plt.figure()
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {name}')
        plt.legend(loc='lower right')
        plt.show()

estimators = [
    ('lr', LogisticRegression(max_iter=1000)),
    ('rf', RandomForestClassifier()),
    ('gb', GradientBoostingClassifier())
]
stacking_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
stacking_model.fit(X_train, y_train)
y_pred_stack = stacking_model.predict(X_test)
y_proba_stack = stacking_model.predict_proba(X_test)

accuracy_stack = accuracy_score(y_test, y_pred_stack)
precision_stack = precision_score(y_test, y_pred_stack, average='macro')
recall_stack = recall_score(y_test, y_pred_stack, average='macro')
f1_stack = f1_score(y_test, y_pred_stack, average='macro')
roc_auc_stack = roc_auc_score(y_test, y_proba_stack, average='macro', multi_class='ovo')

print('Stacking Model Results:')
print(f'Accuracy: {accuracy_stack}')
print(f'Precision: {precision_stack}')
print(f'Recall: {recall_stack}')
print(f'F1-Score: {f1_stack}')
print(f'ROC-AUC: {roc_auc_stack}')
print('\n')
joblib.dump(stacking_model, 'best_model_new.pkl')
```
